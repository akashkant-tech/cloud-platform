```python
import pyttsx3
import webbrowser as wb
#from time import gmtime, strftime
#text1 = open("nms1.txt", "rb").read()
#text2 = text1.decode("utf-8")
engine = pyttsx3.init()
#engine.say('Hi')
engine.runAndWait()

import speech_recognition as sr
r = sr.Recognizer()
mic = sr.Microphone()
with mic as source:
    r.adjust_for_ambient_noise(source, duration=0.5)
    print("say:")
    audio = r.listen(source)
    try:
        text1 = r.recognize_google(audio)
        print(text1)
        engine.say(text1)
        engine.runAndWait()
        if "Sherlock" in text1:
            engine.say("Just say your username and password please")
            print("Just say your username and password please")
            engine.runAndWait()
            print("say :")
            audio = r.listen(source)
            
            try:
                # text = r.recognize_sphinx(audio)
                text = r.recognize_google(audio)
                print(text)
                engine.say(text)
                engine.runAndWait()
                if "Akash" and "redhat" in text:  
                    print("Hi akash ")
                    engine.say("Hi akash")
                    engine.runAndWait()
                #if "21" in text:
                    print("Hi akash now you are authenticated")
                    engine.say("Hi akash now you are authenticated")
                    engine.runAndWait()
                    wb.open_new_tab("http://192.168.43.185/index.html")
                else:
                    print("You are not authenticated")
                    engine.say("You are not authenticated")
                    engine.runAndWait()

            except sr.UnknownValueError:
                print("Google Speech Recognition could not understand audio")

            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))
    except sr.UnknownValueError:
            print("Google Speech Recognition could not understand audio")
    except sr.RequestError as e:
            print("Could not request results from Google Speech Recognition service; {0}".format(e))
```
